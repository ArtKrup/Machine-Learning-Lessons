# Support Vector Machine

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

## Data Loading

```python
cancer = datasets.load_breast_cancer()
```

```python
data = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
data['target'] = cancer.target
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

## Exploratory Data Analysis

```python
sns.heatmap(data.corr())
```

```python
sns.pairplot(data[['mean radius', 'mean concavity', 'mean texture',  'mean smoothness', 'mean compactness', 'target']])
```

## Data Preprocessing

```python
y = data['target']
data = data.drop(columns='target')
X = data
```

## Splitting Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17, stratify=y)
```

## Modeling

```python
svc = SVC(gamma='auto')
```

## Training Model

```python
svc.fit(X_train, y_train)
```

## Predicting Data

```python
predict = svc.predict(X_test)
```

## Evaluating Model

```python
print(classification_report(y_test, predict))
print(confusion_matrix(y_test, predict))
```

## Model Tuning

### Gridsearch

```python
params_grid = {'C': [0.01, 0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001, 0.0001]}
```

```python
grid = GridSearchCV(SVC(), params_grid, refit=True, verbose=2)
```

## Training Parameters

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_, grid.best_score_
```

## Predicting Data

```python
predict = grid.predict(X_test)
```

## Evaluating Model

```python
print(classification_report(y_test, predict))
print(confusion_matrix(y_test, predict))
```
